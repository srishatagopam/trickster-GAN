```python

import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from torchvision import models
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import ToPILImage, Resize, ToTensor
from dataloader import ImageDataset
from models import *
from train import train
from test import test

%reload_ext autoreload
%autoreload 2
```

```python
train_ds = ImageDataset('./data/train/') 
test_ds = ImageDataset('./data/test/')
print(len(train_ds), len(test_ds))
print(train_ds.counts)
print(test_ds.counts)
```

```python
model = models.resnet18(pretrained=True)
num_feat = model.fc.in_features
model.fc = nn.Linear(num_feat, 8) 
model.load_state_dict(torch.load('resnet18_utk.pt'))
```

```python
G = Generator().to('cuda')
D = Discriminator().to('cuda')

G_losses, D_losses = train(G, D, model, './data/train/', device='cuda', 
                           batch_size=32, epochs=20, lim=4096, lr=5e-4)
```

```python
torch.save(G.state_dict(), 'gen_20_4096.pt')
```

```python
def test(G, model, test_dir='./data/test/', lim=None):
    test = ImageDataset(test_dir, lim) if lim is not None else ImageDataset(test_dir)
    
    loader = DataLoader(test, batch_size=32, shuffle=True)
    G.to('cuda')
    model.to('cuda')
    model.eval()
    G.eval()
    test_dict = defaultdict(int, {i:0 for i in range(8)})
    perclass_correct = [0]*8
    perclass_totals = [0]*8
    correct = 0
    total = len(test)

    with torch.no_grad():
          for batch, (image, label) in enumerate(loader):
                image = image.to('cuda')
                pert = Resize(128)(G(image))
                pert = torch.clamp(pert,2, 5)
                label = label.cpu().numpy()
                out = model(image + 5*pert)
                _, preds = torch.max(out, 1)
                preds = preds.detach().cpu().numpy()
#                 print(preds)
                for i,j in zip(preds, label):
                    perclass_totals[j] += 1
                    if i==j:
                        perclass_correct[i] += 1
                correct += np.sum(np.equal(preds, label))
    print(f'Total Acc: {100 * correct / total}%')
    print('Per class Acc:')
    for idx, i in enumerate(test.keys):
        print(f'{test.classes[i]}: {100 * perclass_correct[idx] / perclass_totals[idx]}')
        
```

```python
test(G, model)
```
